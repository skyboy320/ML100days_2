#三種特徵 分類

```python
int_features = []
float_features = []
object_features = []

for dtype, feature in zip(app_train.dtypes, app_train.columns):
    if dtype == 'int64':
        int_features.append(feature)
    elif dtype == 'float':
        float_features.append(feature)
    else:
        object_features.append(feature)
```

計算三種不同類型的特徵的平均數、標準差

```python
app_train[int_features].describe()
app_train[int_features].mean()
```

```python
app_train[float_features].sum()
```

```python
#計算平均數，標準差

print('計算AMT_INCOME_TOTAL 平均數: ', app_train['AMT_INCOME_TOTAL'].mean())
print('計算AMT_INCOME_TOTAL 平均數: ', app_train['AMT_INCOME_TOTAL'].std())
```

```python
#直方圖 方法一
app_train['OWN_CAR_AGE'].plot.hist();
```

```python
#直方圖 方法二
plt.hist(app_train['OWN_CAR_AGE'], color = 'blue')
plt.title('OWN_CAR_AGE')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(app_train['TARGET']);
```

```python
plt.bar(['0','1'], app_train['TARGET'].value_counts());
```

```python
#當處理不為數值型態的特徵時, 可以使用value_counts來計算其各唯一值數量
plt.bar(['F','M','XNA'], app_train['CODE_GENDER'].value_counts())
plt.title('CODE_GENDER')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
